```python
import os
from os import path
import json
from json import loads
import sys
import re
import csv
```

```python
dbRWP = 'rwp352_lingeval_enriched_db.json'

with open(dbRWP) as input_json_file:
    try:
        rwpDB = json.load(input_json_file)
    except json.JSONDecodeError as e:
        print(f"Error decoding JSON in {filename}: {str(e)}")
```

```python

csvDir = 'db_features2csv'
oFile = 'attrStats.csv'
oFilePath = os.path.join(csvDir, oFile)

if not path.exists(csvDir):
   os.makedirs(csvDir)
   print('Created directory ', csvDir)
```

```python

headers = ['exID', 'tgtCnt', 'tgtID', 'patternID', 'pattern', 'rwpCnt', 'pos2verb',
           'optProvCxt', 'provRwp', 'optSdhCxt', 'optSdhRwpCxt', 'asyllRwp', 'hostRwp']
rows = []

for exID in rwpDB['rwp_db']:
    print('processing ... ', exID)
    c_description = rwpDB['rwp_db'][exID]['description']
    c_tgtCnt = c_description['target_count']
    for dKey in c_description.keys():
        if dKey.startswith('d_'+exID+'_p'):
            for qKey in c_description[dKey].keys():
                print(qKey)
                c_q = c_description[dKey][qKey]
                c_exID, c_pID, c_qID = qKey.split('_')[-3:]
                c_pattern = c_q['string']
                c_rwpCnt = c_q['rwp_sequence_length']
                c_syntax = c_q['position_to_verb']
                c_provContext = c_q['optional_prov_sandhi_context']
                c_provRWP = c_q['prov_rwp']
                c_optSandhiContext = c_q['optional_sandhi_context']
                c_optSandhiRwpContext = c_q['optional_rwp_sandhi_context']
                c_hostRWP = c_q['host_rwp']
                c_asyllRWP = c_q['asyllabic_rwp']
                
                c_row = []
                # add exID
                c_row.append(c_exID)
                # add tgtCnt
                c_row.append(c_tgtCnt)
                # add pID/tID
                c_row.append(c_pID)
                # add qID
                c_row.append(c_qID)
                # add pattern
                c_row.append(c_pattern)
                # add rwpCnt
                c_row.append(c_rwpCnt)
                # add c_syntax
                c_row.append(c_syntax)
                # add provContext
                c_row.append(c_provContext)
                # add provRWP
                c_row.append(c_provRWP)
                # add optSandhiContext
                c_row.append(c_optSandhiContext)
                # add c_optSandhiRwpContext
                c_row.append(c_optSandhiRwpContext)
                # add c_asyllRWP
                c_row.append(c_asyllRWP)
                # add c_hostRWP
                c_row.append(c_hostRWP)
                
                rows.append(c_row)

with open(oFilePath, 'w', encoding='utf-8') as csv_file:
    csv_writer = csv.writer(csv_file, lineterminator=os.linesep)
    csv_writer.writerow(headers)
    csv_writer.writerows(rows)

print('DONE')
```
