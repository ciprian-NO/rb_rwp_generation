```python
import os
from os import path
import json
from json import loads
import sys
import re
```

```python
#version = '_3'

iFile = "genDir/genOpt_82.json"

with open(iFile) as input_json_file:
    rwpDB = json.load(input_json_file)

# dir for the cleaned output: cleaned means without the source and notes subelements
oDir = 'clnDir' #+version
oFile = "rule_based_oput_rwp352.json"

if not path.exists(oDir):
   os.makedirs(oDir)

oFilePath = os.path.join(oDir, oFile)
```

```python

for ix, c_id in enumerate(rwpDB['rwp_db']):
    print(str(ix), c_id)
    rwpDB['rwp_db'][c_id].pop('notes')
    rwpDB['rwp_db'][c_id].pop(c_id+'_source')

with open(oFilePath, 'w', encoding='utf-8') as f:
    json.dump(rwpDB, f, ensure_ascii=False, indent=2)

```
